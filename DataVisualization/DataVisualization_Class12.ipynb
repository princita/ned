```python
red_wine = pd.read_csv("winequality-red.csv" ,sep = ";")
white_wine = pd.read_csv("winequality-white.csv",sep = ";")
```

```python
red_corr = red_wine.corr()["quality"][:-1] # red_corr.drop["quality"]
white_corr  = white_wine.corr()["quality"][:-1]
possitive_red = red_corr>0
possitive_white = white_corr>0
```

```python
color_red = possitive_red.map({True:"#3cb371",False:"#ff0000"} )
color_white = possitive_white.map({True:"#3cb371",False:"#ff0000"} )
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import coords
style.use("ggplot")
```

```python
fig , ax = plt.subplots(figsize=(9,5))
ax.barh(white_corr.index,white_corr,label="white",left=2,color=color_white)
ax.barh(red_corr.index,red_corr.values,label="red",color=color_red)
ax.grid(visible = False)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_xticks([])
ax.set_yticks([])

xcoords = coords.x_coords
ycoords = coords.y_coord
for label,x_cord in xcoords.items():
    ax.text(x=x_cord,y=ycoords,s=label)
    ycoords-=1

ax.axvline(0.5,c="grey",alpha=0.1,linewidth=1,ymin=0.1,ymax=0.9)
ax.axvline(1.45,c="grey",alpha=0.1,linewidth=1,ymin=0.1,ymax=0.9)

ax.axhline(-1,xmin=0.01,xmax=0.32,c='grey',linewidth=1,alpha=0.5)
ax.axhline(-1,xmin=0.67,xmax=0.98,c='grey',linewidth=1,alpha=0.5)
ax.text(x=-0.5,y=-1.7,s="-0.5"+" "*37+"+0.5" ,c="grey",alpha=0.5)
ax.text(x=1.5,y=-1.7,s="-0.5"+" "*37+"+0.5" ,c="grey",alpha=0.5)

ax.axhline(11,xmin=0.01,xmax=0.32,c='grey',linewidth=1,alpha=0.5)

ax.axhline(11,xmin=0.67,xmax=0.98,c='grey',linewidth=1,alpha=0.5)

ax.text(x=-0.33,y=11.2,s="RED WINE",weight = "bold")
ax.text(x=1.75,y=11.2,s="WHITE WINE",weight = "bold")

ax.text(x=-0.7,y=13.5,s="Wine quality most storongly coorelted with Alchol",weight = "bold",size=17)
ax.text(x=-0.7,y=12.8,s="Coralaion with the wine quality and its properties",size=13)
ax.text(x=-0.7,y=-2.7,s="Digitech Senegy Pvt Limited"+" "*68+"Ned university",size=12, backgroundcolor="#4d4d4d",color="#f0f0f0")

plt.show()
```

```python
euro = pd.read_csv("euro.csv")
euro
```

```python
euro.columns
```

```python
euro_dollar = euro[[r"Period\Unit:",r"[US dollar ]"]].copy()
euro_dollar
```

```python
euro_dollar.rename(columns={r"Period\Unit:":"Date",r"[US dollar ]":"USD"},inplace =True)
```

```python
euro_dollar
```

```python
euro_dollar.Date = pd.to_datetime(euro_dollar.Date)
euro_dollar
```

```python
euro_dollar.sort_values(by="Date",inplace=True)
```

```python
euro_dollar.reset_index(drop=True,inplace=True)
```

```python
euro_dollar
```

```python
euro_dollar.info()
```

```python
euro_dollar["USD"].value_counts()
```

```python
bol = euro_dollar['USD'] != "-"
```

```python
euro_dollar = euro_dollar[bol]
```

```python
euro_dollar.USD = euro_dollar.USD.astype(float)
```

```python
plt.plot(euro_dollar["Date"],euro_dollar.USD)
plt.show()
```

# moving avarge ,rolling avarage -->moving window rolloing window
Moving Windows

```python
euro_dollar.USD.rolling(7).mean().head(50)
```

```python
rolling = [7,30,50,100,365]
plots = [2,3,4,5,6]
```

```python
plt.figure(figsize=(11,9))
plt.subplot(3,2,1)
plt.plot(euro_dollar["Date"],euro_dollar.USD)
plt.title("Orginal")
for rol,plot in zip(rolling,plots):
    plt.subplot(3,2,plot)
    plt.plot(euro_dollar["Date"],euro_dollar.USD.rolling(rol).mean())
    plt.title(f"Moving Avarge:{rol}")
plt.tight_layout()
plt.show()
```

```python
euro_dollar["Rolling_mean_30"]=euro_dollar.USD.rolling(30).mean()
```
